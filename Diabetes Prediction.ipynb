```python
def classlablepredict(row):
    if row['AGE'] <49:
        age = 'N'
        
    elif row['AGE'] <51:
        age= 'P'
    else:
        age='Y'
    if row['Urea']<3.0:
        urea='N'
    elif row['Urea']<4.0:
        urea='P'
    else:
        urea='Y'
    if row['Cr']<25:
        cr='N'
    elif row['Cr']<30:
        cr='P'
    else:
        cr='Y'
    if row['HbA1c']<5.7:
        hb='N'
    elif row['HbA1c']<6.5:
        hb='P'
    else:
        hb='Y'
    if row['Chol']<2.0:
        ch='N'
    elif row['Chol']<2.5:
        ch='P'
    else:
        ch='Y'
    if row['TG']<1.0:
        tg='N'
    elif row['TG']<2.0:
        tg='P'
    else:
        tg='Y'
    if row['HDL']<0.6:
        hd='N'
    elif row['HDL']<1.6:
        hd='P'
    else:
        hd='Y'
    if row['LDL']<0.8:
        ld='N'
    elif row['LDL']<1.4:
        ld='P'
    else:
        ld='Y'
    if row['VLDL']<0.7:
        vl='N'
    elif row['VLDL']<1.2:
        vl='P'
    else:
        vl='Y'
    if row['BMI']<24.5:
        bmi='N'
    elif row['BMI']<26:
        bmi='P'
    else:
        bmi='Y'
    DS=0
    if age=='Y':
        DS=DS+1
    if urea=='Y':
        DS=DS+1
    if cr=='Y':
        DS=DS+1
    if hb=='Y':
        DS=DS+1
    if ch=='Y':
        DS=DS+1
    if tg=='Y':
        DS=DS+1
    if hd=='Y':
        DS=DS+1
    if ld=='Y':
        DS=DS+1
    if vl=='Y':
        DS=DS+1
    if bmi=='Y':
        DS=DS+1
    if DS>=6:
        DC='Y'
    elif DS>=5:
        DC='P'
    else:
        DC='N'
    return(DC)
```

```python
data1.apply(lambda row: classlablepredict(row), axis = 1)
data1['class']=data1.apply(lambda row: classlablepredict(row),axis =1)
```

```python
def truepositive(row):
    if row["CLASS"]=='Y' and row["class"]=='Y':
        tp=1
        return(tp)
    else:
        tp=0
        return(tp)
```

```python
data1.apply(lambda row: truepositive(row), axis = 1)
data1['truepositive']=data1.apply(lambda row: truepositive(row),axis =1)
```

```python
data1.head(40)
```

```python
def truenegative(row):
    if row["CLASS"]=='N' and row["class"]=='N':
        tn=1
        return(tn)
    else:
        tn=0
        return(tn)
```

```python
data1.apply(lambda row: truenegative(row), axis = 1)
data1['truenegative']=data1.apply(lambda row: truenegative(row),axis =1)
data1.head(40)
```

```python
def falsepositive(row):
    if row["CLASS"]=='N' and row["class"]=='Y':
        fp=1
        return(fp)
    else:
        fp=0
        return(fp)
```

```python
data1.apply(lambda row: falsepositive(row), axis = 1)
data1['falsepositive']=data1.apply(lambda row: falsepositive(row),axis =1)
data1.head(40)
```

```python
def falsenegative(row):
    if row["CLASS"]=='Y' and row["class"]=='N':
        fn=1
        return(fn)
    else:
        fn=0
        return(fn)
```

```python
data1.apply(lambda row: falsenegative(row), axis = 1)
data1['falsenegative']=data1.apply(lambda row: falsenegative(row),axis =1)
data1.head(40)
```

```python
def trueoutcome(row):
    if row["CLASS"]==row["class"]:
        to=1
        return(to)
    else:
        to=0
        return(to)
```

```python
data1.apply(lambda row: trueoutcome(row), axis = 1)
data1['true outcome']=data1.apply(lambda row: trueoutcome(row),axis =1)
data1.head(40)
```

```python
def falsepositiveP(row):
    if (row["CLASS"]=='N' and row["class"]=='P') or (row["CLASS"]=='Y' and row["class"]=='P'):
        fpP=1
        return(fpP)
    else:
        fpP=0
        return(fpP)
```

```python
data1.apply(lambda row: falsepositiveP(row), axis = 1)
data1['falsepositiveP']=data1.apply(lambda row: falsepositiveP(row),axis =1)
data1.head(40)
```

```python
count_tp=data1["truepositive"].value_counts()[1]
count_tn=data1["truenegative"].value_counts()[1]
count_fp=data1["falsepositive"].value_counts()[1]
count_fn=data1["falsenegative"].value_counts()[1]
total = count_tp+count_tn+count_fp+count_fn
accuracy=(count_tp+count_tn)/total *100
print(accuracy)
```
